```python

import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

##**Ingreso total por cada tienda**

```python
ingreso_tienda1 = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

print(f"Ingreso total Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Ingreso total Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Ingreso total Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Ingreso total Tienda 4: ${ingreso_tienda4:,.2f}")
```

##**Ventas por categoría**

```python
print("\n--- Categorías más populares ---")

for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    categorias = (
        df.groupby("Categoría del Producto")
        .size()
        .reset_index(name="Cantidad Vendida")
        .sort_values(by="Cantidad Vendida", ascending=False)
    )
    print(f"\nTienda {i}:\n", categorias)
```

##**Valoración media por tienda**


```python
print("\n--- Calificaciones promedio por tienda ---")
calificacion_tienda1 = tienda["Calificación"].mean()
calificacion_tienda2 = tienda2["Calificación"].mean()
calificacion_tienda3 = tienda3["Calificación"].mean()
calificacion_tienda4 = tienda4["Calificación"].mean()

print(f"Tienda 1: {calificacion_tienda1:.2f}")
print(f"Tienda 2: {calificacion_tienda2:.2f}")
print(f"Tienda 3: {calificacion_tienda3:.2f}")
print(f"Tienda 4: {calificacion_tienda4:.2f}")
```

##**Productos más vendidos y menos vendidos**

```python
print("\n--- Productos más y menos vendidos ---")
for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    productos = (
        df.groupby("Producto")
        .size()
        .reset_index(name="Cantidad Vendida")
        .sort_values(by="Cantidad Vendida", ascending=False)
    )

    mas_vendido = productos.head(1)
    menos_vendido = productos.tail(1)

    print(f"\nTienda {i}:")
    print("Producto más vendido:\n", mas_vendido)
    print("Producto menos vendido:\n", menos_vendido)
```

##**Valor del envío promedio por tienda**

```python
print("\n--- Costo de envío promedio por tienda ---")

for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    costo_prom = df["Costo de envío"].mean()
    print(f"Tienda {i}: ${costo_prom:.2f}")
```

##**Gráficos**

```python
import matplotlib.pyplot as plt

tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

fig, axs = plt.subplots(2, 2, figsize=(14, 10))

for i, (df, nombre) in enumerate(zip(tiendas, nombres_tiendas)):
    top5 = df["Producto"].value_counts().head(5)

    ax = axs[i // 2, i % 2]
    ax.bar(top5.index.astype(str), top5.values)
    ax.set_title(f"Top 5 productos más vendidos - {nombre}")
    ax.set_ylabel("Cantidad vendida")
    ax.set_xlabel("Producto")
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

promedios_calificacion = [df["Calificación"].mean() for df in tiendas]

plt.figure(figsize=(8, 5))
plt.bar(nombres_tiendas, promedios_calificacion)
plt.title("Calificación promedio por tienda")
plt.ylabel("Promedio de calificación")
plt.ylim(0, 5)
plt.tight_layout()
plt.show()

promedios_envio = [df["Costo de envío"].mean() for df in tiendas]

plt.figure(figsize=(8, 5))
plt.bar(nombres_tiendas, promedios_envio)
plt.title("Costo de envío promedio por tienda")
plt.ylabel("Costo promedio de envío")
plt.tight_layout()
plt.show()
```
